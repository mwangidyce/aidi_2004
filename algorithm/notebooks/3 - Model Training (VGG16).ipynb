```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
import numpy as np
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import img_to_array
from numpy import expand_dims
import PIL
```

```python

augmentation=0
while augmentation < 3:
    
    circleDatagen = ImageDataGenerator(horizontal_flip=True,zoom_range=0.1,rescale=1./255)
    CircleTrainPath="../data/images/train/circle/"
    CircleValidationPath="../data/images/test/circle/"
    circleTrain = os.listdir(CircleTrainPath)

    for index, image in enumerate(circleTrain):          # for loop for training images of Circle. 
        filename = image.split('.')
        if(filename[1] == 'png' or filename[1] == 'jpg' or filename[1] == 'jpeg'):
            # load the image using split concept.
            img = load_img(CircleTrainPath+filename[0]+'.'+filename[1])
            
            print(filename)
            
            
            # convert to numpy array
            data = img_to_array(img)
            # expand dimension to one sample
            circleSamples = expand_dims(data, axis=0)

            itCircle = circleDatagen.flow(circleSamples, batch_size=32,save_to_dir= CircleTrainPath,
                              save_prefix="CIRCLE",
                              save_format='png')     #Using datagen.flow, every image is flipped horizontally, 
                                                    #rotated at certain angle and finally stored in a specified directory. 
            batch = itCircle.next()

        #Same concept follows for test images of Square shown below

    squareDatagen = ImageDataGenerator(zoom_range=0.1,rescale=1./255)
    SquareTrainPath="../data/images/train/square/"
    SquareValidationPath="../data/images/test/square/"
    squareTrain = os.listdir(SquareTrainPath)

    for index, image in enumerate(squareTrain):           
        filename = image.split('.')
        if(filename[1] == 'png' or filename[1] == 'jpg' or filename[1] == 'jpeg'):
            img = load_img(SquareTrainPath+filename[0]+'.'+filename[1])
            data = img_to_array(img)
            squareSamples = expand_dims(data, axis=0)


            squareIt = squareDatagen.flow(squareSamples, batch_size=32,save_to_dir= SquareTrainPath,
                              save_prefix="SQUARE",
                              save_format='png')     #Using datagen.flow, every image is flipped horizontally, 
                                                    #rotated at certain angle and finally stored in a specified directory. 
            batch = squareIt.next()
    
    augmentation+=1
```

```python
trainDatagen = ImageDataGenerator(rescale=1./255)
 
testDatagen = ImageDataGenerator(rescale=1./255)
 
train_generator = trainDatagen.flow_from_directory(
    "../data/binary/train",
    target_size=(224,224),
    batch_size=32,
    class_mode='binary')
 
validation_generator = testDatagen.flow_from_directory(
    "../data/binary/test",
    target_size=(224,224),
    batch_size=32,
    class_mode='binary')
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
from tensorflow.keras.optimizers import SGD

#VGG16 Architecture used for Convolution of images

base_model=tf.keras.applications.vgg16.VGG16(
    include_top=False,
    weights=None,
    input_tensor=None,
    input_shape=(224,224,3),
    pooling=None
)

for layer in base_model.layers:
    layer.trainable = False

x=base_model.output

model=Flatten()(x)
model=Dense(units=400,activation="relu")(model)
model=Dense(units=400,activation="relu")(model)
output=Dense(units=1, activation="sigmoid")(model)

model = keras.models.Model(inputs=base_model.input, outputs = output)

from tensorflow.keras.optimizers import Adam
opt = Adam(lr=0.001)
model.compile(optimizer=opt, loss="binary_crossentropy", metrics=['accuracy'])
model.summary()
```

```python
model.fit_generator(steps_per_epoch=1,generator=train_generator,validation_data=validation_generator,validation_steps=1,epochs=100,verbose=1,shuffle=False)
```

```python
_, trainacc = model.evaluate(train_generator, verbose=0)
print("Accuracy")
print('> %.3f' % (trainacc * 100.0))
```

```python
_, validationacc = model.evaluate(validation_generator, verbose=0)
print("Accuracy")
print('> %.3f' % (validationacc * 100.0))     
```
