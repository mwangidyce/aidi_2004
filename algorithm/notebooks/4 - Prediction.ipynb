```python
# make a prediction for a new image.

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.models import load_model
from tensorflow.keras.applications import imagenet_utils
import numpy as np
def load_image(filename):
    #load the image
    img = load_img(filename, grayscale=False, target_size=(224, 224))
    #convert to array
    img = img_to_array(img)
    #reshape into a single sample with 1 channel
    img = img.reshape(1, 224, 224, 3)
    #prepare pixel data
    img = img.astype('float32')
    img = img / 255.0
    return img

CIRCLE="../data/binary/train/circles/Circle0.jpg"
SQUARE="../data/binary/train/square/Square0.jpg"
test1_loc = "../old data/Predict using these images that is stored anywhere/Ball.jpg"


img1 = mpimg.imread(test1_loc)
imgplot = plt.imshow(img1)
plt.show()
img = load_image(test1_loc)
print(img.shape)
model = load_model("../models/shapes_model-2.h5")
# predict the class

prediction = model.predict(img)

#prediction = np.argmax(model.predict(img), axis=-1)

print("Result:",prediction)

test2_loc = "../old data/Predict using these images that is stored anywhere/Biscuit2.jpg"

img1 = mpimg.imread(test2_loc)
imgplot = plt.imshow(img1)
plt.show()
img = load_image(test2_loc)
print(img.shape)

# predict the class

prediction = model.predict(img)

print("Result:",prediction)
```
